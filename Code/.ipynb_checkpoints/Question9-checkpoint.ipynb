```python
from matplotlib . pyplot import *
from math import *
from array import *
from numpy import *
from numpy. random import *
from scipy . misc import *
from scipy . stats import *
from scipy import *
from random import *


#Question 9 : Simulation de Dn
#Convergence p. s .
N=10**4
D=np.zeros(N)
D[0]=0

#Cas pour 1

p =1/2
B=np.random. binomial (1,p,1)
D[1]=D[0]+B

#Cas pour le reste 

for i in range(2,N):
   p =((i-1)-D[i-1])/(i)
   B=np.random. binomial (1,p,1)
   D[i]=D[i-1]+B

T=linspace(1,N ,N)
        
figure(figsize =(20,10))

plot (D,'b')
grid (True)
title ("Convergence presque sure de Dn",fontsize =20,color="blue")

figure(figsize =(20,10))

plot (D/T,'g--')
plot (T,1/2 *linspace (1 ,1 ,N),'r--',lw=3)

grid (True)
title ("Convergence presque sure de Dn/n",fontsize =20,color="blue")
```
