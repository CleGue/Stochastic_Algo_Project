```python
from numpy import *
from random import *
from scipy.stats import binom
import numpy.random as npr
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import scipy.stats
import numpy as np
from scipy.spatial import ConvexHull, convex_hull_plot_2d


def simulate(N):
    tabindice=[]
    G=np.zeros(N+1)
    G[0]=0
    G[1]=0
    G[2]=1
    
    tabindice.append(0)
    tabindice.append(1)
    tabindice.append(2)

    for i in range(3,N+1):
        g2=((i-2)-G[i-1])/i
        g1=2/i
        g0=G[i-1]/i
        
        
        nbaleatoire=random()
        if(nbaleatoire<=g0):
            G[i]=G[i-1]
        ##gauche
        elif(nbaleatoire<=(g0+g1)):
            G[i]=G[i-1]+1
        ##haut
        elif(nbaleatoire<=(g0+g1+g2)):
            G[i]=G[i-1]+2
        tabindice.append(i)
        
    plt.style.use('seaborn')  
    plt.title('Convergence presque sure associe  Gn',fontsize=20,color="blue")
    plt.plot(tabindice, G/tabindice,'darkgreen',lw=3,label="Gn")
    plt.plot(tabindice,2/3 * linspace (1 ,1 ,N+1),'r--',lw=3,label="y=2/3")
    plt.grid (True)
    plt.xlabel('n',color="black",fontsize=15)
    plt.ylabel('Gn/n',color="darkred",fontsize=15)
    plt.legend()
    plt.savefig('Convergence presque sure de Gn n .png')
    plt.show()
    
    plt1.style.use('seaborn')  
    plt1.plot (G,'b',label="Gn")
    plt1.grid (True)
    plt1.title ("Evolution de Gn en fonction de n",fontsize =20,color="blue")
    plt1.xlabel('n',fontsize=15,color="black")
    plt1.ylabel('Gn',fontsize=15,color="darkred")
    plt1.legend()
    plt1.savefig('Convergence presque sure de Gn.png')
    plt1.show()
    
    
    
simulate(10**4)
```
