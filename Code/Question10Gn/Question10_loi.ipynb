```python
from numpy import *
from random import *
from scipy.stats import binom
import numpy.random as npr
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import scipy.stats
import numpy as np
from scipy.spatial import ConvexHull, convex_hull_plot_2d


def simulate(N):
    tabindice=[]
    G=np.zeros(N+1)
    G[0]=0
    G[1]=0
    G[2]=1
    
    tabindice.append(0)
    tabindice.append(1)
    tabindice.append(2)

    for i in range(3,N+1):
        g2=((i-2)-G[i-1])/i
        g1=2/i
        g0=G[i-1]/i
        
        
        nbaleatoire=random()
        if(nbaleatoire<=g0):
            G[i]=G[i-1]
        ##gauche
        elif(nbaleatoire<=(g0+g1)):
            G[i]=G[i-1]+1
        ##haut
        else:
            G[i]=G[i-1]+2
        tabindice.append(i)
        
    return G[N]

tab_loi=[]
        
def simulate2 (repetition,N):
    for i in range (1,repetition):
        l=simulate(N)
        p=math.sqrt(N)*((l/(N)) - (2/3))
        p=math.sqrt(45/8)*p
        tab_loi.append(p)
        
    plt.style.use('seaborn')    
    plt.title('Convergence en loi',fontsize=20,color="blue")
    n, bins, patches = plt.hist(tab_loi, 30, normed=1, facecolor='chartreuse',edgecolor="black",label="densité G")
    y=scipy.stats.norm.pdf(bins, 0, 1)
    plt.plot(bins, y,'r',label="densité loi normale")
    plt.xlabel('Limite',fontsize=15,color="black")
    plt.ylabel('Frequence',fontsize=15,color="darkred")
    plt.legend()
    plt.savefig('Convergence en loi Gn.png')
    plt.show()
        
   
simulate2(4000,4000)
```
