```python
from numpy import *
from random import *
from scipy.stats import binom
import numpy.random as npr
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import matplotlib.mlab as mlab
import numpy as np
import scipy.stats

def simulate(N):
    tab =[]
    D=np.zeros(N+1)
    D[0]=0
    D[1]=0

#Cas pour le reste 

    for i in range(2,(N+1)):
       p =((i-1)-D[i-1])/(i)
       B=np.random. binomial (1,p,1)
       D[i]=D[i-1]+B
    
    return D[N]

tab_loi=[]
    
def simulate2 (repetition,N):
    for i in range (1,repetition):
        l=simulate(N)
        p=math.sqrt(N)*((l/(N)) - (0.5))
        p=math.sqrt(12)*p
        tab_loi.append(p)
        
    plt.style.use('seaborn')
    plt.title('Convergence en loi',fontsize=20,color="blue")
    n, bins, patches = plt.hist(tab_loi, 30, normed=1, facecolor='lightgreen',edgecolor="black",label="densité D")
    y=scipy.stats.norm.pdf(bins, 0, 1)
    plt.plot(bins, y,'r',label="densité loi normale")
    plt.xlabel('Limite',color="black",fontsize=15)
    plt.ylabel('Frequence',color="darkred",fontsize=15)
    plt.legend()
    plt.savefig('Convergence en loi.png')
    plt.show()
        
   
simulate2(4000,4000)
```
