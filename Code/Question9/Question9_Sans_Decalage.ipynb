```python
from matplotlib . pyplot import *
from math import *
from array import *
from numpy import *
from numpy. random import *
from scipy . misc import *
from scipy . stats import *
from scipy import *
from random import *
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1


#Question 9 : Simulation de Dn
#Convergence p. s .
N=10**4
D=np.zeros(N)
D[0]=0
D[1]=0

#Cas pour le reste 

for i in range(2,N):
   p =((i-1)-D[i-1])/(i)
   B=np.random. binomial (1,p,1)
   D[i]=D[i-1]+B

vect_n=linspace(1,N ,N)

plt.style.use('seaborn')
plt.plot (D,'b',label="Dn")
plt.grid (True)
plt.title ("Evolution de Dn en fonction de n",fontsize =20,color="blue")
plt.xlabel('n',fontsize=15,color="black")
plt.ylabel('Dn',fontsize=15,color="darkred")
plt.legend()
plt.savefig('Convergence presque sure de Dn.png')
plt.show()

plt1.style.use('seaborn')
plt1.plot (D/vect_n,'g--',label="D/n")
plt1.plot (vect_n,1/2 *linspace (1 ,1 ,N),'r--',lw=3,label="y=1/2")
plt1.grid (True)
plt1.title ("Convergence presque sure associée à Dn",fontsize =20,color="blue")
plt1.xlabel('n',fontsize=15,color="black")
plt1.ylabel('Dn/n',fontsize=15,color="darkred")
plt1.legend()
plt1.savefig('Convergence presque sure de Dn n.png')
plt1.show()
```
